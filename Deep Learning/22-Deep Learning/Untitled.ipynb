```python
import requests
from bs4 import BeautifulSoup
```

```python
scrap_url = "https://www.youtube.com"
search_url = "/watch?v="
search_hardcode = 'mdFClOESrXM'
```

```python
url = scrap_url + search_url + search_hardcode
```

```python
url
```

```python
sb_get = requests.get(url)
```

```python
sb_get
```

```python
soupeddata  = BeautifulSoup(sb_get.content)
```

```python
soupeddata
```

```python
title = soupeddata.find_all('h1', class_ = 'watch-title-container')
```

```python
type(title)
```
